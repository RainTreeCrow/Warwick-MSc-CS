```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
import scipy

N = 3000
p = 0.2

plt.figure(figsize=(20, 4))

x= list(range(N+1))
pmf_poisson = []
pmf_binomial = []

for k in x:
    res = np.exp(-N*p)
    for i in range(1, k+1):
        res = res * (N*p) / i
    pmf_poisson.append(res)
plt.scatter(x, pmf_poisson, marker=".", color='r', s=10, label='Poisson')

for k in x:
    res = 1
    for i in range(0, k):
        res = res * p * (N-i) / (i+1)
    for i in range(0, N-k):
        res = res * (1-p)
    pmf_binomial.append(res)
plt.scatter(x, pmf_binomial, marker=".", color='b', s=10, label='Binomial')

plt.title(r'$p$='+str(p))
plt.legend()
```
