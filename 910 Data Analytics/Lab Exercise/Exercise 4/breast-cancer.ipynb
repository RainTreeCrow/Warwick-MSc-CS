```python
from distfit import distfit
import numpy as np

X = np.array([24.5] * 1 + [34.5] * 36 + [44.5] * 90 \
    + [54.5] * 96 + [64.5] * 57 + [74.5] * 6)

dfit = distfit()
results = dfit.fit_transform(X)
dfit.plot_summary()
```

```python
dfit = distfit(distr='norm')
dfit.fit_transform(X)
print(dfit.model)
dfit.plot(chart='pdf')
```
